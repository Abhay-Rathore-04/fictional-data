```python
import pandas as pd            #Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Zomato data .csv') #importing file
df
```

```python
df.head(10) #top 10 rows only showing
```

```python
df.info() # information of whole table and its data types
#table has no null values
```

```python
def change_rate(value):
  value = str(value).split('/')
  value = value[0];
  return float(value)

df["rate"] = df["rate"].apply(change_rate)   #making a function to change "/" to float
```

```python
print(df.head()) # rate coloumn values has changed
```

```python
df.info() #info of whole table
```

```python
df.insert(6,"Couples",df['approx_cost(for two people)'])
df.head(4)
#inserting a column name couple with same value as approx_cost clm
```

```python
top_5df = df.sort_values(by = ['rate'], ascending= False).head(6) # making a variable of top_5df which will be used to make graph of top 5 values
```

```python
plt.figure(figsize=(7, 5))

sns.barplot(x='name', y='rate', data=top_5df, palette='Set2')
plt.xticks(rotation=45, ha='right')


plt.title("highest rated restaurants", fontsize=20)
plt.xlabel("Names of Restaurats")
plt.ylabel("Ratings")
plt.show()
```

```python
sns.countplot(x=df['listed_in(type)'],palette='Set2')
plt.xlabel("Type of restaurant")
plt.title("Type of restaurants",fontsize=20) #counted the higest and lowest of types of restaurent
```

```python
sns.countplot(x=df['online_order'],palette='Set2')
plt.title("Order Online")
#counted online order vs offline order
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Couples'], color='lightpink',palette='Set2',bins=10,kde=False)
plt.xlabel('Approximate Cost for couples')
plt.ylabel('Number of Restaurants')
plt.title('Price range is preferred by couples')
plt.show()

# price range preferred by couples graph
```
